### Coding Challenge 1 - Slot Machine game

In this coding challenge for the Python Fundamentals series, you will have to write a slot machine game in Python.

1. Write the `spin_reel()` function that will randomly select three symbols from the `symbols` list. You will need to use the random library randomly select elements from a list.
2. Write the main loop. Inside the loop, ask the player whether they want to continue to play. The loop should be terminated if the player does not want to play again. On every iteration, spin the reel and print the result using the `print_spin()` function.
3. Write the `evaluate_spin()` function to print a message about how much the player has won. Determine yourself how much payout the player should receive for which combination of symbols. Add the `evaluate_spin()` function to the while loop to display the payout after a spin.
4. You can use the `clear_output()` function from the `IPython.display` module in jupyter to clear the output of a code cell on each iteration. Place the `clear_output()` function  within your while loop, so that it will remove the previously printed slot machine result after the players input.



```python
from IPython.display import clear_output
import random

symbols = ['üçí', 'üçä', 'üçã', 'üîî', '‚≠ê', 'BAR', '7']
payouts = {'üçí': 10, 'üçä': 20, 'üçã': 30, 'üîî': 40, '‚≠ê': 50, 'BAR': 100, '7': 50}
# Function to spin the reel
def spin_reel(symbols):
    return random.choices(symbols, k=3)

# Function to evaluate the spin and calculate the payout
def evaluate_spin(spin):
    payout = 0
    unique_symbols = set(spin)
    
    if len(unique_symbols) == 1:
        symbol = spin[0]
        payout = payouts.get(symbol, 0) * 2
        print(f"Jackpot! You've won ${payout}!")
    elif len(unique_symbols) == 2:
        print("Two of a kind!")
        for symbol in unique_symbols:
            if spin.count(symbol) == 2:
                payout = payouts.get(symbol, 0)
    else:
        print("No payout this round.")
        
    return payout

# Function to print the spin result
def print_spin(spin):
    print("+--------------+")
    print("|    CASINO    |")
    print("|--------------|")
    print(f"  {spin[0]:<3} {spin[1]:<3} {spin[2]:<3} ")
    print("|--------------|")
    print("|       [_]    |")
    print("+--------------+")

# Main game loop
def game1():
    play = True
    result = spin_reel(symbols)
    print_spin(result)
    payout = evaluate_spin(result)
    print(f"You've won ${payout}!")
    while play:
        # Ask the player if they want to play again
        play_again = input("Do you want to play again? (yes/no): ").lower()
        play = play_again == 'yes'
        # Spin the reel
        result = spin_reel(symbols)
    
        # Clear previous output before printing new output
        clear_output(wait=True)
        print_spin(result)
    
        # Evaluate the spin result and calculate payout
        payout = evaluate_spin(result) 

    print("Thanks for playing!")
```
